```python
from importMag import *
%matplotlib inline
```

```python
prism = definePrism()
prism.dy = 2.
prism.susc, prism.pinc, prism.pdec = 0.1, 15., 45.
prism.Binc, prism.Bdec = 90, 0
prism.Q, prism.rinc, prism.rdec = 1., 0., 00.,

prism.Bigrf = 5e+4
prism.z0 = -0.5

xr = np.linspace(-2., 2., 20)
yr = np.linspace(-2., 2., 20)
X, Y = np.meshgrid(xr, yr)
Z = np.ones(np.shape(X))

rxLoc = np.c_[X.flatten(order='F'),Y.flatten(order='F'),Z.flatten(order='F')]

bx = getField(prism, rxLoc, "bx","induced")
by = getField(prism, rxLoc, "by","induced")
bz = getField(prism, rxLoc, "bz","induced")
tf = getField(prism, rxLoc, "tf","induced")

fig = plt.figure(figsize=(10,5))
fig = MAG.plot_obs_2D(rxLoc,bx,fig=fig,axs = plt.subplot(1,4,1),varstr = 'Bx')
fig = MAG.plot_obs_2D(rxLoc,by,fig=fig,axs = plt.subplot(1,4,2),varstr = 'By')
fig = MAG.plot_obs_2D(rxLoc,bz,fig=fig,axs = plt.subplot(1,4,3),varstr = 'Bz')
fig = MAG.plot_obs_2D(rxLoc,tf,fig=fig,axs = plt.subplot(1,4,4),varstr = 'Tf')
```
