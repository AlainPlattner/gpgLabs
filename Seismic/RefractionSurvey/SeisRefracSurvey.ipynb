## Interpretation of seismic refraction data

Assume that you have got seismic refraction data, then structure of the earth is unknown and you may want to obtain useful information about the subsurface. To proceed this interpretation of seismic refraction data, let survey region is known as three-layer structure and velocity of each layer is increasing as depth increases. 
Thus, there can be four unknowns:

- v1: velocity of the first layer (m/s)
- v2: velocity of the first layer (m/s)
- v3: velocity of the first layer (m/s)
- z1: depth of the first layer (m)
- z2: depth of the second layer (m)

Based on the above information, we may expect to have four arrivals related to 

- Direct
- Reflected: interface 1
- Refraction: interface 1
- Refraction: interface 2

Below widget will allow you to estimate above unknown information about the subsurface. Parameters for widget are following:

- x0: offset between shot and the first geophone
- dx: spacing between two consecutive geophones
- Fit: checking this activates fittting function (if you click this red line will show up)
- tI: intercept time for a line function (s)
- v: inverse slope of the line function (m/s; which can be velocity of either direct and critically refracted wave)
